```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

# 한글 폰트 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

# 데이터 불러오기
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
cols = ['V1','V2','V3','V4','V5','V6','V7','V8','defect']
df = pd.read_csv(url, names=cols)
```

```python
df['V1'].hist(bins=20, rwidth=0.8)  # 막대 사이 간격 조절
plt.title("센서 V1 값 분포")
plt.xlabel("V1 값")
plt.ylabel("샘플 수")
plt.grid()
plt.show()
```

## Boxplot

```python
sns.boxplot(x='defect', y='V1', data=df)
plt.title("결함 여부에 따른 V1 분포")
plt.xlabel("결함 여부")
plt.ylabel("V1 값")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("센서 간 상관관계 히트맵")
plt.show()
```

```python
df['defect'].value_counts().plot(kind='line')
plt.title("결함 vs 정상 샘플 수")
plt.xlabel("결함 여부 (1=결함, 0=정상)")
plt.ylabel("샘플 수")
plt.grid()
plt.show()
```

```python
df['defect'].value_counts().plot(kind='bar')
plt.title("결함 vs 정상 샘플 수")
plt.xlabel("결함 여부 (1=결함, 0=정상)")
plt.ylabel("샘플 수")
plt.xticks(rotation=0)  # x축 글자 세로로 보이게
plt.grid()
plt.show()
```
